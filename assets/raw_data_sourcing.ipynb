```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["SPY", "NVDA", "TSLA", "XLP", "JNJ"]

# download daily data for the past 12 months
df = yf.download(
    tickers=tickers,
    start="2025-11-05",   # or use pd.Timestamp.today() - pd.DateOffset(years=1)
    end=None,             # defaults to today
    interval="1d",        # daily bars
    group_by='ticker'     # data returned grouped by ticker
)
```

```python
df = df.stack(level=0, future_stack=True).rename_axis(['Date', 'Ticker']).reset_index()
```

```python
# rename columns
df.columns = [c.lower().replace(' ', '_') for c in df.columns]
df.rename(columns={'date': 'trade_date'}, inplace=True)
```
